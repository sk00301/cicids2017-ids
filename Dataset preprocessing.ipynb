```python
#combining csv
import os
import pandas as pd

directory_path = "D:\MachineLearningCSV\MachineLearningCVE"

# List to store DataFrames
dataframes = []

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory_path, filename)
        try:
            # Attempt to read the CSV file
            df = pd.read_csv(file_path, encoding="utf-8", on_bad_lines="skip")  # Skip malformed lines
            dataframes.append(df)
            print(f"Successfully read: {filename}")
        except Exception as e:
            print(f"Error reading {filename}: {e}")

# Combine all DataFrames into one
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    # Save the combined DataFrame to a new CSV file
    output_path = os.path.join(directory_path, "combined_traffic_data.csv")
    combined_df.to_csv(output_path, index=False)
    print(f"Combined CSV saved at: {output_path}")
else:
    print("No valid CSV files to combine.")
```

```python
import pandas as pd
data_path = r"D:\cicids 2017\MachineLearningCVE\combined_Machinelearning.csv"

data = pd.read_csv(data_path)
combined_df = data
data
```

```python
import numpy as np

combined_df.ffill(inplace=True) #forward fill
combined_df.replace([np.inf, -np.inf], np.nan, inplace = True)

combined_df
```

```python
# Check for null values in each column
print(combined_df.isnull().sum())

print(f"Total null values: {combined_df.isnull().sum().sum()}")
```

```python
# Remove rows with any null values
combined_df = combined_df.dropna()

combined_df
```

```python
# Count missing values in each column
print(combined_df.isnull().sum())

print(f"Total missing values: {combined_df.isnull().sum().sum()}")
```

```python
#detecting duplicates 
duplicates_count = combined_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates_count}")
```

```python
#Remove duplicate rows
combined_df = combined_df.drop_duplicates()

print(f"Shape after removing duplicates: {combined_df.shape}")
```

```python
combined_df
```

```python
from scipy import stats
from scipy.stats import zscore
import numpy as np

numeric_columns = combined_df.select_dtypes(include=[np.number]).columns
z_score = np.abs(stats.zscore(combined_df[numeric_columns]))
outliers = (z_score>3).sum(axis=1) < (len(numeric_columns)/2)
combined_df = combined_df[outliers]
combined_df
```

```python
#normalize numeric columns
from sklearn.preprocessing import StandardScaler
 
scaler = StandardScaler()
combined_df[numeric_columns] = scaler.fit_transform(combined_df[numeric_columns])
combined_df
```

```python
import pandas as pd

# Define the full file path where the CSV will be saved
file_path = r'D:/MachineLearningCSV/MachineLearningCVE/Preprocessed dataset.csv'

# Save the DataFrame to the specified location
combined_df.to_csv(file_path, index=False)
print("Saved preprocessed data.")
```
