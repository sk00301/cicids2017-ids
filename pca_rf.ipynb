```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv("D:\\Kartikey\\Project\\.venv\\Preprocessed dataset.csv")

# Extract features and target
X = df.drop(' Label', axis=1)
y = df[' Label']

# Print total number of original features
print(f"Total number of original features: {X.shape[1]}")

# Standardize the features (PCA is sensitive to scaling)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA (keep enough components to explain 95% of the variance)
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Get the number of selected components (features explaining 95% variance)
n_selected_components = X_pca.shape[1]
print(f"Number of selected components (explaining 95% variance): {n_selected_components}")

# Get the loadings (contributions) of each feature to the principal components
loadings = pca.components_

# Get feature names
features = X.columns

# Create a DataFrame for the loadings of the selected components only
pca_contributions = pd.DataFrame(loadings[:n_selected_components], columns=features)

# Calculate the importance of each feature by summing the absolute values of its loadings across the selected components
feature_importance = pca_contributions.abs().sum(axis=0)

# Select features with non-zero importance
selected_features = feature_importance[feature_importance > 0].index.tolist()

# Create importance DataFrame for selected features
importance_df = pd.DataFrame({
    'Feature': selected_features,
    'Importance': feature_importance[selected_features].values
})

# Sort the features by their importance in descending order
importance_df = importance_df.sort_values(by='Importance', ascending=False)

pd.set_option('display.max_rows', None)  # Set this to None to show all rows
pd.set_option('display.max_columns', None)  # Set this to None to show all columns
pd.set_option('display.width', None)  # To ensure the width is sufficient for full display
pd.set_option('display.max_colwidth', None)  # To avoid truncating column content

# Display the selected features
print("\nSelected Features:")
print(", ".join(selected_features))

# Display the number of selected features
print(f"\nNumber of selected features: {len(selected_features)}")

# Display the importance of the selected features
print("\nSelected Feature Importances (Sum of absolute loadings across selected PCs):")
print(importance_df.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the preprocessed dataset
# Replace 'cicids_2017_preprocessed.csv' with the actual path to your dataset
data = pd.read_csv(r'D:\Kartikey\.venv\Preprocessed dataset.csv')

# Split features and target
X = data.drop(columns=[' Label'])  # Replace 'target' with the actual target column name
y = data[' Label']

# Standardize the dataset for PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
pca.fit(X_scaled)

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()

# Select components explaining a threshold of variance (e.g., 95%)
explained_variance = np.cumsum(pca.explained_variance_ratio_)
threshold = 0.95
n_components = np.argmax(explained_variance >= threshold) + 1
print(f"Number of components selected to explain {threshold*100}% variance: {n_components}")

# Perform PCA again with the selected number of components
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# Calculate feature contributions (absolute loadings)
feature_contributions = np.abs(pca.components_).sum(axis=0)
feature_contributions = feature_contributions / feature_contributions.sum()  # Normalize contributions

# Create a DataFrame for feature contributions
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Impact': feature_contributions
}).sort_values(by='Impact', ascending=False)

# Display total number of selected features and their impacts
selected_features = feature_importance.head(n_components)
print(f"Total number of selected features: {len(selected_features)}")
print("Selected features and their impact:")
print(selected_features)

# Plot feature contributions
plt.figure(figsize=(12, 8))
plt.barh(selected_features['Feature'], selected_features['Impact'], color='skyblue')
plt.xlabel('Impact')
plt.ylabel('Feature')
plt.title('Top Features by PCA Impact')
plt.gca().invert_yaxis()
plt.grid()
plt.show()
```
